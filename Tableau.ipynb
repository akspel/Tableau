```python
import pandas as pd
import numpy as np
```

```python
june_df = pd.read_csv("Resources/2019/June_2019.csv")
june_df.head()
```

```python
july_df = pd.read_csv("Resources/2019/July_2019.csv")
july_df.head()
```

```python
august_df = pd.read_csv("Resources/2019/August_2019.csv")
august_df.head()
```

```python
summer_df = pd.concat([june_df, july_df, august_df])
summer_df.head()
```

```python
summer_df.to_csv('Resources/summer_2019.csv', index=False)
```
